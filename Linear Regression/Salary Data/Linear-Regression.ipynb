```python
import numpy as np
import pandas as pd #Data Analysis
import matplotlib.pyplot as plt #Data Visualisation
import seaborn as sns #Data Visualisation
```

```python
data = pd.read_csv("salarydata.csv")
print(data.head(5)) 
```

<h1>Data Visualisation </h1>

```python
sns.scatterplot(x="YearsExperience", y="Salary", data=data, color="blue")
```

```python
plt.scatter(data.YearsExperience,data.Salary, color="red")
```

```python
print(data.corr()) # r = co-efficient as co-relation
```

```python
x=data.iloc[:,0:1] # :is x axis = rows and y axis(after comma) = 0:1 is column
x.head(10)
```

```python
y = data.iloc[:,1:]
y.head(10)
```

```python
# Visualise the data

plt.scatter(x,y)
plt.title("Experience vs Salary")
plt.xlabel("Work Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x,y)
```

<h2><b>Model Prediction</b></h2>

```python
y_prediction = lin.predict(x)
print(y_prediction)
print(y.head(10))
```

```python
plt.scatter(x,y)
plt.plot(x,y_prediction,color ="red")
plt.title("Linear Regression")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
